##Rename Columns: Replace spaces with _ to allow dot notation

```python
housing_df = housing_df.rename(columns = {'TOTAL VALUE ':"TOTAL_VALUE"})
housing_df.columns = [s.strip().replace(' ','_') for s in housing_df.columns]
housing_df.head()
```
