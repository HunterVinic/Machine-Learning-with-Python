```python
import pandas as pd
file_path = "/content/drive/MyDrive/datamining/BostonHousing.csv"
f1_df = pd.read_csv(file_path)
f1_df.head()
```

```python
predictors =['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','RAD','TAX','PTRATIO','LSTAT']
outcome =['CAT. MEDV']
```

```python
scaler.fit(trainData[predictors])

f1Norm = scaler.transform(f1_df[predictors])
f1Norm = pd.DataFrame(f1Norm, columns= ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','RAD','TAX','PTRATIO','LSTAT'])
f1Norm = pd.concat([f1Norm, f1_df[outcome]], axis=1)
f1Norm
```

```python

from sklearn.neighbors import NearestNeighbors

knn = NearestNeighbors(n_neighbors= 5)

f1Norm = f1Norm.iloc[trainData.index]
f1Norm
```

```python
import pandas as pd
file_path = "/content/drive/MyDrive/datamining/Cereals.csv"
f2_df = pd.read_csv(file_path)
f2_df.head()
```

```python
f2_df.fillna(0, inplace=True)
f2_df.head()
```

```python
from sklearn.metrics import pairwise_distances

numerical_data = f2_df.select_dtypes(include =['float64','int64'])
distances = pairwise_distances(numerical_data, metric ='euclidean')

distance_df = pd.DataFrame(distances, index = f2_df['name'], columns =  f2_df['name'])

distance_df.head(10)
```

```python

f2_df_norm =(numerical_data - numerical_data.mean())/numerical_data.std()

distances_scaled = pairwise_distances(f2_df_norm, metric ='euclidean')

scaled_distance_df = pd.DataFrame(distances_scaled, index =f2_df['name'], columns =  f2_df['name'])

scaled_distance_df.head(10)
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
Z= linkage(f2_df_norm, method= 'single')

plt.figure(figsize = (10,7))
dendrogram = dendrogram(Z, labels= f2_df['name'].values, orientation='top', leaf_rotation= 90, leaf_font_size= 10)

plt.title('Dendrogram for Single Linkage')
plt.xlabel('Name of Breakfast Cereals.')
plt.ylabel('Eculidean Distances')
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

 #Perform Hierarchical Clustering using complete Linkage Method
Z= linkage(f2_df_norm, method= 'complete')
#Plot Dendrogram
plt.figure(figsize = (10,7))
dendrogram = dendrogram(Z, labels= f2_df['name'].values, orientation='top', leaf_rotation= 90, leaf_font_size= 10)

plt.title('Dendrogram for Complete Linkage')
plt.xlabel('Name of Breakfast Cereals')
plt.ylabel('Eculidean Distances')
plt.show()
```

### The longest line that has not been crossed by any other lines is the 4th line from the top. Therefore the number of clusters should be 9-10.
