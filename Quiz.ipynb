```python
import pandas as pd
file_path = "/content/drive/MyDrive/datamining/LaptopSalesJanuary2008.csv"
laptop_df = pd.read_csv(file_path)
laptop_df.head()
```

```python
laptop_df = laptop_df.rename(columns ={'Store Postcode': 'Store_Postcode'})
laptop_df.head()
```

```python
laptop_df = laptop_df.rename(columns ={'Retail Price': 'Retail_Price'})
laptop_df.head()
```

```python
ax = sns.barplot(data=laptop_df, x="Store_Postcode", y="Retail_Price")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
plt.show()
```

```python
ax = sns.barplot(data=laptop_df, x="Store_Postcode", y="Retail_Price")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
average = np.mean(laptop_df["Retail_Price"])
ax.axhline(average, color='r', linestyle='--', label='Average')

plt.legend()
plt.show()
```

```python
ax = sns.boxplot(x= "Store_Postcode", y = 'Retail_Price', data = laptop_df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
plt.xlabel('Store_Postecode',fontsize = 12)
plt.ylabel('Retail_Price', fontsize = 12)
plt.title('Distribution of Store by Retail Price', fontsize = 14)
plt.show()
```

```python
import pandas as pd
file_path = "/content/drive/MyDrive/datamining/LaptopSales.csv"
sales_df = pd.read_csv(file_path)
sales_df.head()
```

```python
sales_df = sales_df.rename(columns ={'Retail Price': 'Retail_Price'})
laptop_df.head()
```

```python
average = sales_df["Retail_Price"].mean()
print("Average Retail Price:", average)
```

```python
corr = sales_df.corr()
sns.heatmap(corr, vmin = -1, vmax = 1, cmap ='RdBu')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,4))
sns.heatmap(corr,annot = True, fmt = "0.1f", cmap = 'RdBu', center = 0, ax= ax, vmin = -1, vmax = 1)
plt.show()
```
